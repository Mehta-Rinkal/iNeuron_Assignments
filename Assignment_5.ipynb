3. What is the most significant distinction between a dictionary and a list?

A dictionary and a list are two distinct data structures in Python. Lists are ordered collections of elements, accessible by integer indexes, while dictionaries store key-value pairs, allowing data retrieval by unique keys. Lists hold only values, which can be of any data type, whereas dictionaries contain key-value pairs with keys typically being strings or immutable data types. Lists are ordered, maintaining the sequence of elements, whereas dictionaries are unordered, so the order of key-value pairs may vary. The choice between lists and dictionaries depends on the data organization and access requirements of your program.

```python
# List
my_list = [1, 2, 3, 4]

# Dictionary
my_dict = {'name': 'John', 'age': 30, 'city': 'New York'}
```

5. If a dictionary is stored in spam, what is the difference between the expressions 'cat' in spam and 'cat' in spam.keys()?

- 'cat' in spam: This checks if 'cat' is a key in the dictionary spam. It only looks at the keys and doesn't consider the values associated with those keys.

- 'cat' in spam.keys(): This explicitly checks if 'cat' is in the list of keys generated by the keys() method of the dictionary spam. It's essentially the same as the first expression but makes it explicit that you are checking the keys.

8. How do you "pretty print" dictionary values using which module and function?

To "pretty print" dictionary values in Python, you can use the pprint module, which stands for "pretty-print." Specifically, you can use the pprint function from this module, which is designed to format and display complex data structures, such as dictionaries and lists, in a more human-readable and visually organized way.

```python
import pprint

my_dict = {
    'name': 'rinkal',
    'age': 25,
    'city': 'Ahmedabad',
    'hobby': ['reading', 'coocking', 'dancing'],
    'contact': {
        'email': 'rinkal@gmail.com',
        'phone': '555-555-5555'
    }
}

pp = pprint.PrettyPrinter(indent=2)  # Create a PrettyPrinter object with an indentation of 4 spaces
pp.pprint(my_dict)
```
